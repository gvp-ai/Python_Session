#1. break Statement
The break statement is used to terminate a loop prematurely. When break is executed, control exits the loop immediately, and the program continues with the next statement after the loop

```python
for num in range(10):
    if num == 5:
        break
    print(num)
```

#2. continue Statement
The continue statement skips the remaining code in the current iteration of a loop and moves to the next iteration. It does not terminate the loop but skips to the next cycle.

```python
for num in range(5):
    if num == 3:
        continue
    print(num)
```

```python
##Skipping even numbers
x = 0
while x < 10:
    x += 1
    if x % 2 == 0:
        continue
    print(x)
```

#3. pass Statement
The pass statement is a null operation. It is used as a placeholder where syntactically some code is required but no action is desired.

```python
for num in range(5):
    if num == 3:
        pass  # Placeholder for future logic
    else:
        print(num)
```

#4.Match statement
The match statement, introduced in Python 3.10, implements structural pattern matching. It allows developers to match patterns in data structures (like tuples, lists, dictionaries, etc.) and perform actions based on those patterns. The match statement is similar to a switch or case statement found in other languages but is more powerful due to its pattern-matching capabilities.

#Syntax
match subject:
    case pattern_1:
        # Action for pattern_1
    case pattern_2:
        # Action for pattern_2
    case _:
        # Default action (optional)


```python
x="_"
match x:
    case 1:
        print("It's one")
    case 2:
        print("It's two")
    case _:
        print("Unknown value")
```

```python
day="Monday"
match day:
    case "Monday":
        print("Start of the week")
    case "Friday":
        print("Weekend is near")
    case _:
        print("A regular day")

```

```python
# Get user input
day_number = int(input("Enter a day number (1-7): "))

# Match statement to print the corresponding day
match day_number:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case 6:
        print("Saturday")
    case 7:
        print("Sunday")
    case _:
        print("Invalid day number! Please enter a number between 1 and 7.")
```

#Variable-Length Arguments
These allow passing an arbitrary number of arguments to a function.

***args**: Used for a variable number of positional arguments.
****kwargs**: Used for a variable number of keyword arguments.

```python
def print_all(*args):
    for item in args:
        print(item)

print_all(1, 2, 3, 4)  # Output: 1 2 3 4
```

```python
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=30)
# Output:
# name: Alice
# age: 30
```

**Local Scope**: Variables defined inside a function are only accessible within that function.
**Global Scope**: Variables defined outside of any function are accessible globally throughout the program.

```python
x = 10  # Global variable

def my_function():
    x = 5  # Local variable
    print("Inside function:", x)

my_function()
print("Outside function:", x)
```

#Recursion
A function that calls itself is known as a recursive function. Recursion is useful for problems that can be broken down into smaller subproblems, such as computing factorials or Fibonacci numbers

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(5))  # Output: 120
```

#Closures
A closure occurs when a function is defined inside another function and can capture variables from its enclosing scope.

```python
def outer():
    x = 10
    def inner():
        print(x)
    return inner

closure_function = outer()
closure_function()  # Output: 10
```

**2. Tuple**

A tuple is an immutable sequence type in Python. It is used to store collections of items that should not change throughout the program. Tuples are similar to lists, but with a key difference: tuples cannot be modified after creation (immutable).

**Key Features**

Ordered (elements have a specific sequence).

Immutable (elements cannot be changed after creation).

Allows duplicate elements.

**Syntax**

tuple_name = (element1, element2, element3, ...)




#Creating tuples


```python
# Empty tuple
empty_tuple = ()

# Tuple with one element (comma is required)
single_element_tuple = (42,)  # Note the trailing comma

# Tuple with multiple elements
numbers = (1, 2, 3, 4)

# Heterogeneous tuple
mixed_tuple = (1, "hello", 3.14, True)

# Nested tuple
nested_tuple = ((1, 2), (3, 4))
```

#Immutability of Tuples
Tuples cannot be modified directly. Attempts to change elements, add elements, or remove elements will result in errors.

```python
colors = ("red", "green", "blue")

# Attempting to change an element
colors[1] = "yellow"  # TypeError: 'tuple' object does not support item assignment

# Attempting to append an element
colors.append("yellow")  # AttributeError: 'tuple' object has no attribute 'append'
```

#Set Operations
Python sets support common mathematical operations.

**Union (| or union())**:
Returns all unique elements from both sets.

```python
a = {1, 2, 3}
b = {3, 4, 5}
print(a | b)  # Outputs: {1, 2, 3, 4, 5}
print(a.union(b))  # Outputs: {1, 2, 3, 4, 5}
```

**Intersection (& or intersection())**:
Returns elements common to both sets.

```python
print(a & b)  # Outputs: {3}
print(a.intersection(b))  # Outputs: {3}
```

**Difference (- or difference()):**
Returns elements in the first set but not in the second.

```python
print(a - b)  # Outputs: {1, 2}
print(a.difference(b))  # Outputs: {1, 2}
```

**Symmetric Difference (^ or symmetric_difference()):**
Returns elements in either set but not in both.

```python
print(a ^ b)  # Outputs: {1, 2, 4, 5}
print(a.symmetric_difference(b))  # Outputs: {1, 2, 4, 5}
```

```python
##removing duplicates
nums = [1, 2, 2, 3, 4, 4]
unique_nums = set(nums)
print(unique_nums)  # Outputs: {1, 2, 3, 4}
```

**4. Dictionary**

A dictionary in Python is a collection data type that stores data in key-value pairs. It is highly versatile and widely used for data organization and manipulation.

**Key Features**

Unordered (as of Python 3.7, dictionaries maintain insertion order).

Mutable (values can be updated).

Keys must be unique and immutable (e.g., strings, numbers).

**Syntax**

dictionary_name = {key1: value1, key2: value2, ...}


##Creating dictionary

```python
# Empty dictionary
empty_dict = {}

# Dictionary with data
person = {"name": "Alice", "age": 25, "city": "New York"}

# Using dict() constructor
person_dict = dict(name="Bob", age=30, city="London")

# Dictionary with mixed data types
mixed_dict = {"name": "Charlie", "is_student": True, "scores": [90, 85, 88]}
```

By using keys

```python
person = {"name": "Alice", "age": 25}
print(person["name"])  # Outputs: Alice
```

By using the get method

```python
print(person.get("age"))  # Outputs: 25
print(person.get("gender", "Not Specified"))  # Outputs: Not Specified
```

#Adding key value pairs

```python
erson = {"name": "Alice", "age": 25}
person["age"] = 26  # Update value
person["gender"] = "Female"  # Add new key-value pair
print(person)
```

#Removing Key-Value Pairs:
**pop(key)**: Removes and returns the value for the key.



```python
person = {"name": "Alice", "age": 25, "gender": "Female"}
content=person.pop("age")  # Removes "age"
print("Content of pop :",content)
print(person)
```

**del**: Deletes a key-value pair.

```python
del person["gender"]
print(person)
```

**popitem()**: Removes and returns the last inserted key-value pair.

```python
person={"name":"Alice","age":25}
print(person.popitem())
print(person)
```

**clear()**: Removes all items.

```python
person.clear()
print(person)
```

```python
##Priniting keys
person = {"name": "Alice", "age": 25, "gender": "Female"}
for key in person:
    print(key)  # Outputs: name, age
```

```python
##Printing values
for value in person.values():
    print(value)  # Outputs: Alice, 25
```

```python
for key, value in person.items():
    print(f"{key}: {value}")  # Outputs: name: Alice, age: 25
```

```python
students = {
    "Alice": {"age": 25, "city": "New York"},
    "Bob": {"age": 30, "city": "London"},
}

# Access nested dictionary
print(students["Alice"]["city"])  # Outputs: New York
```

```python
person = {"name": "John", "age": 30, "city": "New York"}

print(person["name"])  # Accessing a value: Output -> John

person["age"] = 31  # Modifying a value
print(person)  # Output -> {'name': 'John', 'age': 31, 'city': 'New York'}

person["gender"] = "Male"  # Adding a new key-value pair
print(person)  # Output -> {'name': 'John', 'age': 31, 'city': 'New York', 'gender': 'Male'}

del person["city"]  # Removing a key-value pair
print(person)  # Output -> {'name': 'John', 'age': 31, 'gender': 'Male'}
```
